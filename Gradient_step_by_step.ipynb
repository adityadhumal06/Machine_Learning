```python
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples=4,n_features =1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
prediction = reg.predict(x)
prediction
```

```python
plt.scatter(x,y)
plt.plot(x,prediction,color='red')
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0

# y = mx + b

y_pred = ((78.35 * x) + 0).reshape(4) # because total 4 points
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,prediction,color='red',label='OLS')
plt.plot(x,y_pred,color='green',label='Gradient b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
```

```python
loss_slope = -2 * np.sum(y-m*x.ravel() - b)
loss_slope
```

```python
# Lets take learning rate = 0.1

lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
# calculating the new_intercept

b = b - step_size
b
```

```python
y_pred1 = ((78.35*x) + b).reshape(4)
y_pred1
```

```python
plt.scatter(x,y)
plt.plot(x,prediction,color='black',label= 'OLS')
plt.plot(x,y_pred, color='green',label='b={}'.format(b))
plt.plot(x,y_pred1,color='red',label='b=0')
plt.legend()
plt.show()
```

```python
# iteration 2

loss_slope = -2 * np.sum(y-m*x.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 *x + b)).reshape(4)
y_pred2
```

```python
plt.scatter(x,y)
plt.plot(x,prediction,color='black',label='OLS')
plt.plot(x,y_pred2,color='green',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='pink',label='b = {}'.format(b))
plt.plot(x,y_pred,color='blue',label='b=0')
plt.legend()
plt.show()
```

```python
# using loop

b = -100
m = 78.35
lr = 0.1

epochs = 100

for i in range(epochs):
  loss_slope = -2*np.sum(y-m*x.ravel() - b)
  b = b - (lr*loss_slope)
  y_pred = m*x+b
  plt.plot(x,y_pred)
plt.scatter(x,y)
```
